```python
def experience_avg(job):
    filtered_df = df[df['title_description'] == job].copy()
    filtered_df['agency_start_year'] = filtered_df['agency_start_date'].str[:4]
    filtered_df['total_years'] = filtered_df['fiscal_year'].astype(int) - filtered_df['agency_start_year'].astype(int)
    filtered_df['experience'] = filtered_df['total_years']  
    
    entry = filtered_df['experience'] <= 5
    midlevel = (5 < filtered_df['experience']) & (filtered_df['experience'] <= 9)
    senior = filtered_df['experience'] > 9
    
    filtered_df['experience'] = filtered_df['experience'].astype(str)
    
    filtered_df.loc[entry, "experience"] = "Entry"
    filtered_df.loc[midlevel, "experience"] = "Midlevel"
    filtered_df.loc[senior, "experience"] = "Senior"

    entry_avg = filtered_df[filtered_df['experience'] == 'Entry']['regular_gross_paid'].mean()
    midlevel_avg = filtered_df[filtered_df['experience'] == 'Midlevel']['regular_gross_paid'].mean()
    senior_avg = filtered_df[filtered_df['experience'] == 'Senior']['regular_gross_paid'].mean()    
        
    return entry_avg, midlevel_avg, senior_avg

experience_avg('IT INFRASTRUCTURE ENGINEER')
```

```python
lat = []
lon = []
for borough in df['work_location_borough']:
    if borough == 'MANHATTAN':
        lat.append(40.7831)
        lon.append(-73.9712)
    elif borough == 'BROOKLYN':
        lat.append(40.6782)
        lon.append(-73.9442)
    elif borough == 'BRONX':
        lat.append(40.8448)
        lon.append(-73.8648)
    elif borough == 'QUEENS':
        lat.append(40.7282)
        lon.append(-73.7949)
    else:
        lat.append(None)
        lon.append(None)
df['lat'] = lat
df['lon'] = lon

df
```

```python
apikey = 'fca_live_2aV9QQzkSxecs8YJjI4wkZKNWemXLTPhQ1RFG5Ae'
def get_exchange_rates(currency, pay, apikey):
    url = "https://api.freecurrencyapi.com/v1/latest"
    querystring = {"apikey": apikey}
    headers = {"X-API-KEY": apikey}
    
    try:
        response = requests.get(url, params=querystring, headers=headers)
        response.raise_for_status()
        data = response.json()
        
        exchange_rate = data['data'].get(currency)
        
        if exchange_rate is not None:
            final_exchange = pay * exchange_rate
            return final_exchange
        
    except Exception as e:
        print(f"Error fetching exchange rates: {e}")
        return None
get_exchange_rates('AUD',100000,'fca_live_2aV9QQzkSxecs8YJjI4wkZKNWemXLTPhQ1RFG5Ae')
```

```python
job = sorted(list(df['title_description'].unique())) 
currencies = ['*Original*','AUD', 'BGN', 'BRL', 'CAD', 'CHF', 'CNY', 'CZK', 'DKK', 'EUR', 'GBP', 'HKD', 'HRK', 'HUF', 'IDR', 'ILS', 'INR', 'ISK', 'JPY', 'KRW', 'MXN', 'MYR', 'NOK', 'NZD', 'PHP', 'PLN', 'RON', 'RUB', 'SEK', 'SGD', 'THB', 'TRY', 'USD', 'ZAR']

display(HTML("<h1>NYC's Best Tech Careers</h1>"))
display(HTML("Let us help you get the best pay, location, and potential earnings for your tech career in NYC!"))
display(HTML("Select the job and the currency you want to convert potential earnings to and we will do the rest!"))

@interact_manual(jobs=job, currency=currencies)   
def onclick(jobs, currency):
    if jobs:
        filtered_df = df[df['title_description'] == jobs]
        pay = filtered_df['regular_gross_paid'].mean()
        max_pay_borough = filtered_df.groupby('work_location_borough')['regular_gross_paid'].mean().idxmax()
        avg_pay_max_borough = filtered_df[filtered_df['work_location_borough'] == max_pay_borough]['regular_gross_paid'].mean()

        entry_avg, midlevel_avg, senior_avg = experience_avg(jobs)
        exchange_rate = get_exchange_rates(currency, pay, apikey)
        
        if exchange_rate is not None:
            converted_pay = exchange_rate
        else:
            print("")
            converted_pay = pay
        
        m = folium.Map(location=[40.7128, -74.0060], zoom_start=10)
        
        filtered_data = filtered_df[filtered_df['work_location_borough'] == max_pay_borough]
        for index, row in filtered_data.iterrows():
            folium.Marker(location=[row['lat'], row['lon']], popup=f"{row['work_location_borough']} is the highest paid borough for the {jobs} job", icon=folium.Icon(color='green', icon="briefcase")).add_to(m)
        
        display(HTML("<h1>Pay Analysis</h1>"))
        display(HTML(f"<li>The average regular gross pay for {jobs} is ${pay:.2f}."))
        display(HTML(f"<li>The average regular gross pay converted to {currency} currency is ${converted_pay:.2f}."))
        display(HTML(f"<li>The best borough to work in for this career is {max_pay_borough}."))
        display(HTML(f"<li>The average regular gross pay in the best borough is ${avg_pay_max_borough:.2f}."))
        display(HTML(f"<h3>Average pay for a {jobs} after: </h3>"))
        display(HTML(f"<li> Less than 5 years of experience: ${entry_avg:.2f}. "))
        display(HTML(f"<li> 5 to 9 years of experience: ${midlevel_avg:.2f}. "))
        display(HTML(f"<li> More than 9 years of experience: ${senior_avg:.2f}."))
        if senior_avg >= 100000:
            display(HTML(f"<h1>This career has great earning potential!"))
        display(sns.barplot(data=filtered_df, x='title_description',y='regular_gross_paid', hue='work_location_borough', errorbar=None))
        display(HTML(f"<h3>Borough with the highest earnings on average </h3>"))
        display(m)
        display(HTML(f"<h3>Comparison of boroughs earnings: </h3>"))
    else:
        pay = df['regular_gross_paid'].mean()
        print(f"The overall average regular gross paid is {pay}")
```
